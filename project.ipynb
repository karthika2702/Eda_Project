#### **    Identifying the mode of payment most preferred

```python
pivot = pd.pivot_table(
    data=data,
    index='Payment_method',
    values='Customer_satisfaction'
)
print(pivot)
# this data indicates that customers prefer cash payyment over other types of payment methods 
```

```python
pivot = pd.pivot_table(
    data=data,
    index='Payment_method',
    values='Customer_satisfaction'
).reset_index()

plt.figure(figsize=(8, 4))  
sns.barplot(data=pivot, x='Customer_satisfaction', y='Payment_method', palette='Set2')
plt.xlabel('Customer Satisfaction')
plt.ylabel('Payment Method')
plt.title('Average Customer Satisfaction by Payment Method')
plt.show()

```

### The above chart clearly indicates that the Customers prefer cash payment over other modes of payment

### **   Identifying the food item sold more

```python
pivot = pd.pivot_table(
    data=data,
    index='Item',
    values='Total'
)

print(pivot)
#Burgers are sold more
```

```python


pivot = pd.pivot_table(
    data=data,
    index='Item',
    values='Total'
).reset_index()

plt.figure(figsize=(5, 6))  


sns.barplot(data=pivot, x='Item', y='Total',color='grey')
plt.xlabel('Item')
plt.ylabel('Total')
plt.title('Total by Item')

plt.xticks(rotation=45)  
plt.tight_layout()  

plt.show()
```

### Burgers are sold more when compared to other items and drinks are sold the least.

### ** categorising items based on their selling price 

```python
pivot = pd.pivot_table(
    data=data,
    index='Item',
    values='Price'
)
print(pivot)
```

```python



pivot = pd.pivot_table(
    data=data,
    index='Item',
    values='Price'
)


plt.figure(figsize=(4, 6))
sns.barplot(x=pivot.index, y="Price", data=pivot, palette="Set2")
plt.xlabel("Item")
plt.ylabel("Price")
plt.title("Price Distribution by Item")
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()

```

### Burgers are the most expensive item in the shop

### **  Do weather conditions play a role in the customer eating pattern


```python
weather_data = data.groupby('Weather')['Total'].sum().reset_index()


plt.figure(figsize=(8, 8))
plt.pie(weather_data['Total'], labels=weather_data['Weather'], autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('total sales by Weather')
plt.show()
```

### weather conditions do affect the eating habit of people. During sunny days customers prefer to eat more. The chart indicates that about 50% of a shop's transactions takesplace during sunny days

###  **Do special offers affect the sales attitude

```python
offer_data = data.groupby('Special_offers')['Total'].sum().reset_index()


plt.figure(figsize=(8, 8))
plt.pie(offer_data['Total'], labels=offer_data['Special_offers'], autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Total sales by Offers')
plt.show()
```

### It is clear that the special offers provided by the shop dont have any significance over the selling behavior.People tend to buy more even during the non offer times.

### **  Identify the busiest transaction time in the canteen

```python



pivot = pd.pivot_table(
    data=data,
    index='Time',
    values='Total'
).reset_index()

plt.figure(figsize=(7, 6)) 
sns.barplot(data=pivot, x='Time', y='Total', palette='husl')  
plt.xlabel('Time')
plt.ylabel('Total')
plt.title('Total sales by Time')
plt.show()
```

### The study shows that the peak time, where most of the transactions takesplace in the said shop is at 2'O' clock
